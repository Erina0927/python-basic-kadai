```python
class Human:
      # コンストラクタの定義
      def __init__(self, name, age):
        self.name = name
        self.age = age

      # メソッドの定義
      def print_info(self):
        print(f"私の名前は{self.name}です。年齢は{self.age}です。")

# Humanクラスをインスタンス化
human = Human("侍花子",38)

# 属性にアクセスして値を出力
human.print_info()
```
