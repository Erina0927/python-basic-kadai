```python
# Humanクラスの定義
class Human():
  # コンストラクタの定義
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # メソッドの定義
  def check_adult(self):
    if self.age > 20:
      print(f"{self.age}歳以上は大人です")
    else:
      print(f"{self.age}歳は大人ではありません")

# インスタンス生成
human1 = Human("侍一郎", 40)
human2 = Human("侍二郎", 30)
human3 = Human("侍三郎", 20)
human4 = Human("侍四郎", 10)
human5 = Human("侍五郎", 5)

# リストへ追加
list = [human1,human2,human3,human4,human5]

# リストの数分だけcheck_adultメソッド呼び出し
for human_age in list:
  human_age.check_adult()
```
